## Q1. Apply regular expression for form validation. Create your domain-form using TkinterModule.
#### ➔ Form should contain Text box [For Name, Email Id, Phone number], Dropdown [for Gender], Spinbox [for Year/DoB] and other necessary widgets required for your domain.
#### ➔ Validate Your Name, Email Id, Phone number in the form.

```python
import tkinter as tk
import re

# Function to validate email using regular expression
def validate_email(email):
    email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    return re.match(email_pattern, email)

# Function to validate phone number using regular expression
def validate_phone(phone):
    phone_pattern = r'^\d{10}$'
    return re.match(phone_pattern, phone)

# Function to validate name (letters only, at least 2 characters)
def validate_name(name):
    name_pattern = r'^[A-Za-z ]{2,}$'
    return re.match(name_pattern, name)

# Function to handle form submission
def submit_form():
    name = name_entry.get()
    email = email_entry.get()
    phone = phone_entry.get()
    gender = gender_var.get()
    dob = dob_var.get()

    if not validate_name(name):
        result_label.config(text="Invalid Name", fg="red")
    elif not validate_email(email):
        result_label.config(text="Invalid Email", fg="red")
    elif not validate_phone(phone):
        result_label.config(text="Invalid Phone Number", fg="red")
    else:
        result_label.config(text="Form Submitted Successfully", fg="green")

# Create the main window
root = tk.Tk()
root.title("Form Validation")
root.geometry("350x450")  

# Name
name_label = tk.Label(root, text="Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Email
email_label = tk.Label(root, text="Email:")
email_label.pack()
email_entry = tk.Entry(root)
email_entry.pack()

# Phone Number
phone_label = tk.Label(root, text="Phone Number:")
phone_label.pack()
phone_entry = tk.Entry(root)
phone_entry.pack()

# Gender
gender_label = tk.Label(root, text="Gender:")
gender_label.pack()
gender_options = ["Male", "Female", "Other"]
gender_var = tk.StringVar(root)
gender_var.set(gender_options[0])  # Default value
gender_dropdown = tk.OptionMenu(root, gender_var, *gender_options)
gender_dropdown.pack()

# Year/Date of Birth
dob_label = tk.Label(root, text="Year/Date of Birth:")
dob_label.pack()
dob_var = tk.StringVar(root)
dob_spinbox = tk.Spinbox(root, from_=1900, to=2023, textvariable=dob_var)
dob_spinbox.pack()

# Submit Button
submit_button = tk.Button(root, text="Submit", command=submit_form)
submit_button.pack()

# Result Label
result_label = tk.Label(root, text="", fg="green")
result_label.pack()

root.mainloop()
```
