## Q2. Perform the Exploratory Data Analysis on your domain-based dataset and demonstrate the retrieved insights using “Matplotlib” modules. Visualize hidden insights using appropriate plots (graphs) [Usage of line plot and scatter plot are mandatory]

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the dataset (replace with your dataset)
df = pd.read_csv("C:/Users/nikhi/Desktop/Python/Lab_g09_2/Date_Fruit_Datasets/Date_Fruit_Datasets.csv")

# Summary statistics
summary = df.describe()

# Line plot - Age distribution
plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.hist(df['StdDevRR'], bins=20, color='skyblue', edgecolor='black')
plt.title('Standard Deviations')
plt.xlabel('StdDevRR')
plt.ylabel('StdDevRG')

# Scatter plot - Income vs. Spending
plt.subplot(1, 4, 2)
plt.scatter(df['AREA'], df['PERIMETER'], color='coral', alpha=0.7)
plt.title('AREA vs. PERIMETER')
plt.xlabel('AREA')
plt.ylabel('PERIMETER')

# Bar graph - Education level counts
plt.subplot(1, 4, 3)
education_counts = df['MAJOR_AXIS'].value_counts()
education_labels = education_counts.index
plt.bar(education_labels, education_counts, color='lightgreen')
plt.title('Education Level Counts')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Exclude non-numeric columns before calculating correlation
numeric_columns = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()

# Plot the correlation heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

# Display the plot
plt.show()




# Display summary statistics
print("Summary Statistics:")
print(summary)

# Display the plots
plt.tight_layout()
plt.show()
```
